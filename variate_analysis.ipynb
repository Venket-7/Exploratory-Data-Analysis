```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

1. UNIVARIATE VISUALIZATIONS (ONE VARIABLE)

1️⃣ Histogram (Distribution of Numerical Data)

📌 Purpose

Shows frequency distribution

Helps identify skewness & outliers

```python
data = pd.DataFrame({"Age":[22,25,30,22,40,35,25,28,32,45]})

plt.hist(data["Age"], bins=5)
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Age Distribution")
plt.show()
```

🧠 Explanation

bins=5 → number of bars
Height of bars = count of values in that range

2️⃣ Box Plot (Outlier Detection)

📌 Purpose

Detects outliers

Shows median, quartiles, IQR

```python
plt.boxplot(data["Age"])
plt.title("Age Box Plot")
plt.ylabel("Age")
plt.show()
```

🧠 Explanation
Center line → Median
Dots outside whiskers → Outliers

3️⃣ Bar Chart (Categorical Data)

📌 Purpose

Compare category counts

```python
gender = pd.DataFrame({"Gender":["M","F","M","F","M","M","F"]})
gender["Gender"].value_counts().plot(kind="bar")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.title("Gender Count")
plt.show()
```

🧠 Explanation
X-axis → categories
Y-axis → frequency

4️⃣ Count Plot (Better Bar Chart – Seaborn)

📌 Purpose

Simplest way to plot categorical data

```python
sns.countplot(x="Gender", data=gender)
plt.title("Gender Distribution")
plt.show()
```

2. BIVARIATE VISUALIZATIONS (TWO VARIABLES)

5️⃣ Scatter Plot (Numerical vs Numerical)

📌 Purpose

Shows relationship & correlation

```python
df = pd.DataFrame({
    "Area":[800,1000,1200,1500,1800],
    "Price":[50,65,80,120,150]
})

plt.scatter(df["Area"], df["Price"])
plt.xlabel("Area (sqft)")
plt.ylabel("Price (Lakhs)")
plt.title("Area vs Price")
plt.show()
```

🧠 Explanation

Upward trend → positive correlation

Random → no correlation

6️⃣ Line Plot (Trend Over Time)

📌 Purpose

Used for time-series data

```python
sales = pd.DataFrame({
    "Month":[1,2,3,4,5],
    "Sales":[200,250,300,280,350]
})

plt.plot(sales["Month"], sales["Sales"], marker='o')
plt.xlabel("Month")
plt.ylabel("Sales")
plt.title("Monthly Sales Trend")
plt.show()
```

7️⃣ Box Plot (Numerical vs Categorical)

📌 Purpose

Compare distributions across categories

```python
df = pd.DataFrame({
    "Gender":["M","F","M","F","M","F"],
    "Salary":[50,60,55,65,70,75]
})

sns.boxplot(x="Gender", y="Salary", data=df)
plt.title("Salary by Gender")
plt.show()
```

8️⃣ Bar Plot (Mean Comparison)

📌 Purpose

Shows average value per category

```python
sns.barplot(x="Gender", y="Salary", data=df)
plt.title("Average Salary by Gender")
plt.show()
```

📈 7️⃣ Grouped Bar Chart (Category vs Category)

```python
data = pd.DataFrame({
    "Gender":["Male","Female","Male","Female"],
    "Department":["IT","IT","HR","HR"],
    "Count":[40,30,35,25]
})
pivot = data.pivot(index="Department", columns="Gender", values="Count")
pivot.plot(kind="bar")
plt.xlabel("Department")
plt.ylabel("Count")
plt.title("Grouped Bar Chart")
plt.show()
```

3. MULTIVARIATE VISUALIZATIONS (3+ VARIABLES)

9️⃣ Pair Plot (Best Multivariate Plot)

📌 Purpose

Shows all pairwise relationships

Used in ML EDA heavily

```python
iris = sns.load_dataset("iris")
sns.pairplot(iris, hue="species")
plt.show()
```

🧠 Explanation

Diagonal → univariate

Off-diagonal → bivariate

🔟 Heatmap (Correlation Matrix)

📌 Purpose

Detect multicollinearity

Feature selection

```python
plt.figure(figsize=(8,6))

numeric_iris = iris.select_dtypes(include=['number'])

sns.heatmap(numeric_iris.corr(), annot=True, cmap="coolwarm")

plt.title("Feature Correlation Heatmap")
plt.show()
```

1️⃣1️⃣ 3D Scatter Plot (Advanced)

📌 Purpose

Visualize 3 numerical features

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(iris["sepal_length"],
           iris["sepal_width"],
           iris["petal_length"])

ax.set_xlabel("Sepal Length")
ax.set_ylabel("Sepal Width")
ax.set_zlabel("Petal Length")
plt.show()
```

1️⃣2️⃣ Violin Plot (Distribution + Density)

📌 Purpose

Combines box plot + KDE

```python
sns.violinplot(x="species", y="sepal_length", data=iris)
plt.title("Sepal Length Distribution by Species")
plt.show()
```
